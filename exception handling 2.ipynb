 ### Q2. Write a python program to print Python Exception Hierarchy.

```python
def print_exception_hierarchy (exception_class,level=0):
    print(" "*level +exception_class.__name__)
    for subclass in exception_class.__subclasses__():
        print_exception_hierarchy(subclass,level+1)
```

```python
print_exception_hierarchy(Exception)
```

 ### Q4. Why LookupError class is used? Explain with an example KeyError and IndexError.

The lookup error is a base class is used to handle when the keyerror and index error getting fail to handle the program and it is basically used in the dictionary and the list

Key error occured when a key is not represented in the dictionary

```python
d={"a":1,"b":2}
try:
    print(d["c"])
except KeyError as e:
    print("keyerror:",e)
```

Index error occured when the list of index is out of range

```python
l=[1,2,3]
try:
    print(l[3])
except IndexError as e:
    print("Indexerror",e)
```
